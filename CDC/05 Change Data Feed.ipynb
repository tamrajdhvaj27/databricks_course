```python
%sql
CREATE TABLE dev.bronze.delta_demo_pepole(first_name STRING, age LONG) using delta TBLPROPERTIES(delta.enableChangeDataFeed = TRUE)
```

```python

df = spark.createDataFrame([
    ("Bob",23),
    ("Alice",25),
    ("Charlie",27),
    ("David",29)
]).toDF("first_name","age")


display(df)
```

```python
df.write.mode('append').format('delta').saveAsTable("dev.bronze.delta_demo_pepole")
```

```python
%sql
select * from dev.bronze.delta_demo_pepole
```

```python
%sql
delete from dev.bronze.delta_demo_pepole where first_name = 'Bob';
```

```python
%sql
update dev.bronze.delta_demo_pepole set age = 30 where first_name ='Alice'
```

```python
df_cdc = spark.read.format("delta")\
    .option("readChangeFeed","true")\
    .option("startingVersion",0)\
    .table("dev.bronze.delta_demo_pepole")
display(df_cdc)
```
