```python
%sql
create or replace table dev.bronze.sales_partitioned(
  order_id    string,
  order_date  date,
  country     string,
  category string,
  amount double
)

using delta
partitioned by(country,category)
```

```python
%sql
insert into dev.bronze.sales_partitioned values
('1', '2024-06-01', 'USA', 'electronics',1200.50),
('2', '2024-06-02', 'Canada', 'electronics',1500.75),
('3', '2024-06-03', 'USA', 'clothing',800.25),
('4', '2024-06-04', 'Canada', 'clothing',1000.00)
```

```python
df = spark.read.table("dev.bronze.sales_partitioned")
df.write.partitionBy("country",'category').save('/Volumes/banking/source/volume/sales_partitioned')
```
