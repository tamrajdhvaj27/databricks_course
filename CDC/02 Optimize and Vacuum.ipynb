```python
from pyspark.sql import Row

data = [
    Row(id =1, name = 'Alice', age=30),
    Row(id =2, name = 'Bob', age=20),
    Row(id =3, name = 'Charlie', age=50),
    Row(id =4, name = 'David', age=40)
    ]

df = spark.createDataFrame(data)

df.write.format('delta').mode('overwrite').saveAsTable('dev.bronze.delta_demo')

delta_df = spark.read.table('dev.bronze.delta_demo')

display(delta_df)
```

```python
df.write.format('delta').mode('append').save('/Volumes/banking/source/volume/mytable')
```
